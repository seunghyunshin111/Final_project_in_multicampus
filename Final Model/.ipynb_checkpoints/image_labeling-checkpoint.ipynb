```python
import io
import os
import pandas as pd
from google.cloud import vision
```

```python
# API 인증을 위한 json 파일을 윈도우 환경 변수로 지정
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = 'c:/secret_key/My Project 45072-ed8deab854f2.json'
```

```python
# 저장해 둔 여행지 이미지를 불러오기
path_dir = 'c:/final_project_practice/positano'
file_list = os.listdir(path_dir)
```

```python
# 희망하는 휴양지 키워드 입력
score_df = pd.DataFrame()

for i in range(len(file_list)):
    
    # 이미지를 차례대로 읽어줌
    file_name = os.path.join(path_dir, file_list[i])
    with io.open(file_name, 'rb') as image_file:
        content = image_file.read()
        
    # 불러온 이미지를 vision api의 label detection에 넣어줌
    client = vision.ImageAnnotatorClient()
    image = vision.types.Image(content=content)
    response = client.label_detection(image=image)
    labels = response.label_annotations
    
    # label detection으로 얻은 10개의 label에서 원하는 항목이 있으면 점수를 더해줌
    score = 0
    for label in labels:
        if label.description in ['Sea', 'Cliff', 'estate']: # 희망하는 키워드
            score += 10
        elif label.description in ['Sky', 'Vacation']:
            score += 5
    temp = pd.DataFrame({'Image': file_list[i], 'Score': score}, index=[i])
    score_df = score_df.append(temp)
```

#### 라벨_딕셔너리 형태로 판다스 데이터프레임 만들기

```python
# 1) 이미지마다 labels 출력
for i in range(len(file_list)):
    
    # 이미지를 차례대로 읽어줌
    file_name = os.path.join(path_dir, file_list[i])
    with io.open(file_name, 'rb') as image_file:
        content = image_file.read()
        
    # 불러온 이미지를 vision api의 label detection에 넣어줌
    client = vision.ImageAnnotatorClient()
    image = vision.types.Image(content=content)
    response = client.label_detection(image=image, max_results=100)
    labels = response.label_annotations
    print(len(labels))
    print(labels)
```

```python
# 2) 여행지(place_list), 라벨(label_list), 라벨 길이(len_list) 각각의 리스트 만들기
place_list = []
label_list = []
len_list = []

for i in range(len(file_list)):
    
    # 이미지를 차례대로 읽어줌
    file_name = os.path.join(path_dir, file_list[i])
    with io.open(file_name, 'rb') as image_file:
        content = image_file.read()
        
    # 불러온 이미지를 vision api의 label detection에 넣어줌
    client = vision.ImageAnnotatorClient()
    image = vision.types.Image(content=content)
    response = client.label_detection(image=image, max_results=100)
    labels = response.label_annotations
#     print(len(labels))
#     print(labels)
    
    print(file_list[i])
    place_list.append(file_list[i].replace(".jpg","").replace(".JPG","").replace(".jpeg","").replace(".gif","").replace(".PNG","").replace(".png",""))
    # 여기까지하면 파일 이름 출력 OK
    
    temp = {} # label 임시 저장소인 temp 딕셔너리를 생성
    
    for label in labels:
        temp[label.description] = round(label.score, 2) # score: 소수점 2자리에서 반올림 전처리
    
    len_list.append(len(temp))
#     print(file_list[i])
    label_list.append(temp)
```

```python
# 3) 각각의 리스트 출력
print(place_list) # 여행지 이름 리스트
print(len(label_list)) # 라벨 갯수
print(label_list) # 여행이미지별 라벨 리스트
print(len_list) # 라벨별 길이 리스트
```

```python
# 4) 각각의 리스트를 바탕으로 판다스 데이터프레임 만들기
total_df = pd.DataFrame({'place': place_list, 'label': label_list, 'len_label': len_list})
total_df
```

#### 라벨_레이블 형태로 판다스 데이터프레임 만들기

```python
# 1) 라벨_딕셔너리를 판다스 데이터프레임으로 변환
label_table_raw = pd.DataFrame.from_dict(label_list)
label_table_raw
```

```python
# 2) 여행지, 라벨_딕셔너리를 넣을 새로운 판다스 데이터프레임 생성
total_len = len(label_table_raw.columns)*len(place_list)
print(total_len)

label_table = pd.DataFrame(columns=['place', 'label', 'score'], index=list(range(0,total_len,1)))
label_table
```

```python
# 3) 새로 만든 판다스 데이터프레임에 차근차근 넣기
k=0
for i in range(len(place_list)):
    for j in range(len(label_table_raw.columns)):
        label_table.iloc[k+j,0] = place_list[i]
        label_table.iloc[k+j,1] = label_table_raw.columns[j]
        label_table.iloc[k+j,2] = label_table_raw.iloc[i,j]
    k = j*(i+1)
    
print(label_table)
```

```python
# 4) 결측치 제거한 판다스 데이터프레임 완성
table = label_table.dropna(axis=0)
table
```
